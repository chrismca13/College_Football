```python
# Read in data for years 15 - 19
# We exclude 20 because it was the weird covid year
# 2021 requires extra data cleaning, columns are formatted differently
years = ['15', '16', '17', '18', '19']

# Establish empty, base dataframe:
df_15 = pd.read_csv('cfb15.csv')
df_all = df_15.iloc[0:0, :]
df_all['year'] = ''


# Read in all 5 years of data and union together into one big df (df_all)
for i in years:
    
    file = 'cfb{}.csv'.format(i)
    df = pd.read_csv(file)
    
    df['year'] = i
    
    df_all = pd.concat([df_all, df])
    
df_all.shape   
```

```python
# Data manipulation for 2021 year:
    # (a) split out the win and loss column
    # (b) remove index column, remove column that we split from step (a)
    
df_21 = pd.read_csv('cfb21.csv')

# Split out win-loss column into 2 columns
df_21[['Win', 'Loss']] = df_21['Win-Loss'].str.split('-', 1, expand=True)
```
